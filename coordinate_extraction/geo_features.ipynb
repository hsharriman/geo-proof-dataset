```python
import re
# Find all occurrences of Name(val1, val2, val3)
pattern = r"([A-Z])\(([^,]+),\s*([^,]+),\s*([^)]+)\)"
matches = re.findall(pattern, gpt_output)

pt = {}
for match in matches:
    name, val1, val2, val3 = match
    # convert coordinates to float, keep label as string
    pt[(round(float(val1),2), round(float(val2),2))] = name

print(pt)
```

```python
from combine_geometry import get_on_lines_list
on_line = get_on_lines_list(lines, points)
on_line
```

```python
def get_label(point, pt_dict):
    x,y = point
    x = round(x,2)
    y = round(y,2)
    return pt_dict[(x,y)]
```

```python
for line, point in on_line:
    line_start = point_conversion_dict[tuple(line[:2])]
    line_end = point_conversion_dict[tuple(line[2:])]
    point = point_conversion_dict[tuple(point)]

    line_start_label = get_label(line_start, pt)
    line_end_label = get_label(line_end, pt)
    point_label = get_label(point, pt)

    print(f"on_line({line_start_label}{line_end_label}, {point_label})")
    
```
